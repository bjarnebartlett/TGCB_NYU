```python
# ---
# Assignment: Modeling Glycolysis Reactions
# Course: TGCB 2026
# Bjarne Bartlett Demonstration
# ---

import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

print("Libraries loaded successfully.")

# INTRODUCTION
# In this assignment, we model the first step of glycolysis: 
# the phosphorylation of Glucose by Hexokinase. 
# We will compare the Law of Mass-Action, Michaelis-Menten, and Hill Kinetics.

```

```python
# ---------------------------------------------------------
# PART 1: MASS ACTION KINETICS
# ---------------------------------------------------------
# Reaction: G (Glucose) + A (ATP) -> P (G6P) + D (ADP)
# Rate = k * [G] * [A]

# Function defining the ODEs
def mass_action_system(y, t, k):
    G, A, P = y
    dGdt = -k * G * A
    dAdt = -k * G * A
    dPdt =  k * G * A
    return [dGdt, dAdt, dPdt]

# Parameters & Initial Conditions
k_val = 0.1
# Initial state: [Glucose=10.0, ATP=5.0, G6P=0.0]
initial_ma = [10.0, 5.0, 0.0] 
t = np.linspace(0, 50, 100)

# Run Simulation
sol_ma = odeint(mass_action_system, initial_ma, t, args=(k_val,))

# TEST PRINT
print("--- Mass Action Test ---")
print(f"Initial Glucose: {sol_ma[0, 0]}")
print(f"Final Glucose (t=50): {sol_ma[-1, 0]:.4f}")
print(f"Final G6P Produced: {sol_ma[-1, 2]:.4f}")

```

```python
# ---------------------------------------------------------
# PART 2: MICHAELIS-MENTEN KINETICS
# ---------------------------------------------------------
# Assuming ATP is in excess, the rate is dependent on [G]
# V = (Vmax * [G]) / (Km + [G])

# Function for MM Kinetics
def michaelis_menten_system(G, t, Vmax, Km):
    dGdt = -(Vmax * G) / (Km + G)
    return dGdt

# Parameters
Vmax_val = 1.0
Km_val = 2.0
G0 = 10.0

# Run Simulation
sol_mm = odeint(michaelis_menten_system, G0, t, args=(Vmax_val, Km_val))

# TEST PRINT
print("--- Michaelis-Menten Test ---")
print(f"Initial Glucose: {G0}")
print(f"Remaining Glucose at t=50: {sol_mm[-1][0]:.4f}")

```

```python
# ---------------------------------------------------------
# PART 3: HILL EQUATION (Cooperativity)
# ---------------------------------------------------------
# Modeling Phosphofructokinase (PFK) which shows sigmoidal behavior
# V = (Vmax * [G]^n) / (K^n + [G]^n)

# Function for Hill Kinetics
def hill_system(G, t, Vmax, K, n):
    dGdt = -(Vmax * (G**n)) / (K**n + (G**n))
    return dGdt

# Parameters
n_low = 1   # No cooperativity
n_high = 4  # High cooperativity (like the Repressilator)

# Run Simulations
sol_hill_low = odeint(hill_system, G0, t, args=(Vmax_val, Km_val, n_low))
sol_hill_high = odeint(hill_system, G0, t, args=(Vmax_val, Km_val, n_high))

# TEST PRINT
print("--- Hill Kinetics Test ---")
print(f"Glucose at t=10 (n=1): {sol_hill_low[20][0]:.4f}") # index 20 is approx t=10
print(f"Glucose at t=10 (n=4): {sol_hill_high[20][0]:.4f}")
print("Note: If n=4 is lower than n=1 at t=10, the 'switch' is working faster.")
```

```python
# Use a consistent time array for all models
t = np.linspace(0, 50, 100) 
G0 = 10.0
Vmax_val = 1.0
Km_val = 2.0
n = 4 # Define n for use in labels

# Part 1: Mass Action
sol_ma = odeint(mass_action_system, [10.0, 5.0, 0.0], t, args=(0.1,))

# Part 2: Michaelis-Menten
sol_mm = odeint(michaelis_menten_system, G0, t, args=(Vmax_val, Km_val))

# Part 3: Hill Kinetics
sol_hill_low = odeint(hill_system, G0, t, args=(Vmax_val, Km_val, 1))
sol_hill_high = odeint(hill_system, G0, t, args=(Vmax_val, Km_val, n))

# ---------------------------------------------------------
# VISUALIZATION - PLOT 1: MASS ACTION
# ---------------------------------------------------------
plt.figure(figsize=(8, 5))
plt.plot(t, sol_ma[:, 0], label='Glucose (Mass Action)', color='blue')
plt.plot(t, sol_ma[:, 2], label='G6P (Mass Action)', linestyle='--', color='orange')
plt.title('Mass Action Kinetics', fontsize=14)
plt.xlabel('Time (s)')
plt.ylabel('Concentration')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

# ---------------------------------------------------------
# VISUALIZATION - PLOT 2: ENZYME KINETICS
# ---------------------------------------------------------
plt.figure(figsize=(8, 5))
plt.plot(t, sol_mm, label='Michaelis-Menten (n=1)', color='green')
# FIXED: Changed sol_hill to sol_hill_high to resolve NameError
plt.plot(t, sol_hill_high, label=f'Hill Kinetics (n={n})', color='red', linewidth=2)
plt.title('Enzyme Saturation & Cooperativity', fontsize=14)
plt.xlabel('Time (s)')
plt.ylabel('Glucose Concentration')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

# ---------------------------------------------------------
# VISUALIZATION - PLOT 3: THE "SHOULDER" COMPARISON
# ---------------------------------------------------------
plt.figure(figsize=(10, 6))
plt.plot(t, sol_ma[:, 0], label='Mass Action (G + ATP)', color='gray', linestyle='--')
plt.plot(t, sol_mm, label='Michaelis-Menten (n=1)', color='blue', linewidth=2)
plt.plot(t, sol_hill_high, label=f'Hill Kinetics (n={n})', color='red', linewidth=2)

plt.title('Comparison of Glycolysis Kinetic Models', fontsize=14)
plt.xlabel('Time (seconds)', fontsize=12)
plt.ylabel('Glucose Concentration [G]', fontsize=12)
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

print("Plotting complete. Compare the 'shoulder' of the red line (n=4) to the others.")
```
